The filesystem consists of a tree of files (plain data) and directories (which can contain other directories or files). The outermost directory is called /. You can navigate around the filesystem, moving into or out of directories and listing the contents of the directory you're currently in.

Within the terminal output, lines that begin with $ are commands you executed, very much like some modern computers:

cd means change directory. This changes which directory is the current directory, but the specific result depends on the argument:
cd x moves in one level: it looks in the current directory for the directory named x and makes it the current directory.
cd .. moves out one level: it finds the directory that contains the current directory, then makes that directory the current directory.
cd / switches the current directory to the outermost directory, /.
ls means list. It prints out all of the files and directories immediately contained by the current directory:
123 abc means that the current directory contains a file named abc with size 123.
dir xyz means that the current directory contains a directory named xyz.

Here, there are four directories: / (the outermost directory), a and d (which are in /), and e (which is in a). These directories also contain files of various sizes.

Since the disk is full, your first step should probably be to find directories that are good candidates for deletion. To do this, you need to determine the total size of each directory. The total size of a directory is the sum of the sizes of the files it contains, directly or indirectly. (Directories themselves do not count as having any intrinsic size.)

The total sizes of the directories above can be found as follows:

The total size of directory e is 584 because it contains a single file i of size 584 and no other directories.
The directory a has total size 94853 because it contains files f (size 29116), g (size 2557), and h.lst (size 62596), plus file i indirectly (a contains e which contains i).
Directory d has total size 24933642.
As the outermost directory, / contains every file. Its total size is 48381165, the sum of the size of every file.
To begin, find all of the directories with a total size of at most 100000, then calculate the sum of their total sizes. In the example above, these directories are a and e; the sum of their total sizes is 95437 (94853 + 584). (As in this example, this process can count files more than once!)

Find all of the directories with a total size of at most 100000. What is the sum of the total sizes of those directories?

```python
sampleIn = """$ cd /
$ ls
dir a
14848514 b.txt
8504156 c.dat
dir d
$ cd a
$ ls
dir e
29116 f
2557 g
62596 h.lst
$ cd e
$ ls
584 i
$ cd ..
$ cd ..
$ cd d
$ ls
4060174 j
8033020 d.log
5626152 d.ext
7214296 k"""
```

```python
commands = sampleIn.splitlines(keepends=False)
print(commands)
```

```python
from typing import List
import os

# hacked from chris at https://github.com/CWallis92/advent-of-code-2022/blob/main/solutions/day-7.ts

with open("input.txt", "r") as f:
    commands = f.read().strip().split("\n")

class TreeNode():
    def __init__(self, key: str, size: int = 0, parent = None) -> None:
        self.key: str = key
        self.size: int = size
        self.parent = parent
        self.children = []
    
    def isLeaf(self) -> bool:
        return len(self.children) == 0
    
    def hasChildren(self) -> bool:
        return len(self.children) > 0
    
    def updateSize(self, size: int) -> None:
        self.size = size

class Tree():

    def __init__(self, key: str, size: int = 0) -> None:
        self.root = TreeNode(key, size)
    
    def preOrderTraversal(self, node = None) -> None:
        if node is None:
            node = self.root
            # print("node is", node.key)

        yield node
        
        if len(node.children) > 0:
            for child in node.children:
                yield from self.preOrderTraversal(child)
    
    def postOrderTraversal(self, node = None) -> None:
        if node == None:
            node = self.root
        
        if len(node.children) > 0:
            for child in node.children:
                yield from self.postOrderTraversal(child)
        
        yield node
    
    def insert(self, parentNodeKey: str, key: str, size: int = 0) -> bool:
        for node in self.preOrderTraversal():
            if node.key == parentNodeKey:
                # print(parentNodeKey, node.key)
                node.children.append(TreeNode(key, size, node))
                return True
        return False
    
    def remove(self, key: str) -> None:
        for node in self.preOrderTraversal():
            filtered = list(filter(lambda x: x.key != key, node.children))
            if len(filtered) != len(node.children):
                node.children = filtered
                return True
        return False
    
    def find(self, key: str) -> TreeNode:
        for node in self.preOrderTraversal():
            # print(node.key)
            if node.key == key:
                return node
        return None
    
    def updateSizes(self) -> None:
        for node in self.postOrderTraversal():
            total = 0
            if node.isLeaf():
                continue
            for child in node.children:
                total += child.size
            node.updateSize(total)
        return False


    def getTotalOverLimit(self) -> int:
        total = 0
        for node in self.preOrderTraversal():
            if node.size < 100000 and not node.isLeaf():
                total += node.size
        return total
    
    def smallestDeletableTableSize(self, totalSpace=70_000_000, unusedSpace=30_000_000) -> int:
        spaceToDelete = unusedSpace - (totalSpace - self.root.size)
        return spaceToDelete
    
    def getTablesToDelete(self, totalSpace = 70000000, unusedSpace = 30000000) -> List[int]:
        deleteTables = []

        smallestDeleteTableSize = self.smallestDeletableTableSize(totalSpace, unusedSpace)

        for node in self.preOrderTraversal():
            if node.size >= smallestDeleteTableSize and not node.isLeaf():
                deleteTables.append(node.size)
        
        return deleteTables

tree: Tree = Tree("/")

currentNode: TreeNode = tree.find("/")

for command in commands:

    ops = command.split(" ")

    if ops[1] == "ls":
        continue

    if command.startswith("$"):
        if ops[2] == "/":
            currentNode = tree.find("/")
        elif ops[2] == "..":
            currentNode = currentNode.parent
        else:
            currentNode = tree.find(
                os.path.join(currentNode.key, ops[2])
            )
    else:
        
        size, name = ops[0], ops[1]

        try:
            size = int(size)
        except:
            size = 0
        tree.insert(currentNode.key, os.path.join(currentNode.key, name), size)
    
    tree.updateSizes()


print(1)
print(tree.getTotalOverLimit())

print(2)
print(min(tree.getTablesToDelete()))
```
